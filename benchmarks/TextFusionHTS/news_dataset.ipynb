```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
# Load the CSV file
df = pd.read_csv('news_avg_embs.csv')
df.head()
```

```python
df.info()
```

```python
# Define the start date and time interval
start_date = datetime(2015, 11, 1)
time_interval = timedelta(minutes=20)

# Generate timestamps for each row
timestamps = [start_date + i * time_interval for i in range(len(df))]

# Insert the 'Date' column as the first column
df.insert(0, 'Date', timestamps)

# Convert the data from format 'YYYY-MM-DD HH:MM:SS' to 'MM-DD-YYYY'
df['Date'] = pd.to_datetime(df['Date']).dt.strftime('%m-%d-%Y')

df.head()
```

```python
# Save the modified CSV
df.to_csv('news_avg_embs.csv', index=False)
```
